```python
dfs = [df12,df13,df14,df15,df16,df17,df18,df19,df20]
df = pd.concat(dfs ,ignore_index = True)
```

```python
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()/len(df)*100
```

```python
columns_to_drop = ["DistrictHeating_RenewableEmissionFactor",
                  "DistrictCooling_RenewableEmissionFactor"]
df.drop(columns = columns_to_drop,inplace = True)
```

```python
#pd.set_option('display.max_columns', None)
df.isnull().sum()/len(df)*100
```

```python
df["Sub Sector"] = df["Sub Sector"].bfill()
df["City"] = df["City"].fillna(df["City"].mode()[0])
df["Postal Code"] = df["Postal Code"].ffill()
df["Unit of Measure"] = df["Unit of Measure"].fillna(df["City"].mode()[0])
df["Annual Flow (M)"] = df["Annual Flow (M)"].fillna(df["Annual Flow (M)"].mode()[0])
df["Electricity_Quantity"]  = df["Electricity_Quantity"].fillna(df["Electricity_Quantity"].mean())
df["Electricity_Unit"] = df["Electricity_Unit"].fillna(df["Electricity_Unit"].mode()[0])
df["NaturalGas_Quantity"]  = df["NaturalGas_Quantity"].fillna(df["NaturalGas_Quantity"].mean())
df["NaturalGas_Unit"] = df["NaturalGas_Unit"].fillna(df["NaturalGas_Unit"].mode()[0])
df["FuelOil12_Quantity"]  = df["FuelOil12_Quantity"].fillna(df["FuelOil12_Quantity"].mean())
df["FuelOil46_Quantity"]  = df["FuelOil46_Quantity"].fillna(df["FuelOil46_Quantity"].mean())
df["Propane_Quantity"]  = df["Propane_Quantity"].fillna(df["Propane_Quantity"].mean())
df["Coal_Quantity"]  = df["Coal_Quantity"].fillna(df["Coal_Quantity"].mode()[0])
df["Wood_Quantity"]  = df["Wood_Quantity"].fillna(df["Wood_Quantity"].mean())
df["DistrictHeating_Quantity"]  = df["DistrictHeating_Quantity"].fillna(df["DistrictHeating_Quantity"].mean())
df["DistrictCooling_Quantity"]  = df["DistrictCooling_Quantity"].fillna(df["DistrictCooling_Quantity"].mean())
df["Number of Portables"]  = df["Number of Portables"].fillna(df["Number of Portables"].mean())
df["GHG Emissions KG"] = df["GHG Emissions KG"].fillna(df["GHG Emissions KG"].median())
df["Energy Intensity ekWh_sqft"] = df["Energy Intensity ekWh_sqft"].fillna(df["Energy Intensity ekWh_sqft"].median())
df["Energy Intensity ekWh_mega_litre"] = df["Energy Intensity ekWh_mega_litre"].fillna(df["Energy Intensity ekWh_mega_litre"].median())
df["Energy Intensity GJ_m2"] = df["Energy Intensity GJ_m2"].fillna(df["Energy Intensity GJ_m2"].median())
df["Energy Intensity GJ_mega_litre"] = df["Energy Intensity GJ_mega_litre"].fillna(df["Energy Intensity GJ_mega_litre"].median())
```
